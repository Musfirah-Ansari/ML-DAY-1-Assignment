# ***Simple Banking System***

> *Welcome to The "Simple Bank" - Your Personal Money Management System*


```python
# ML DAY 1 Assignment By "MUSFIRAH FAYYAZ"

accountsDictionary = {}

def createNewAccount():
    print("\n*** Create a New Account ***\n")
    completeName = input("Enter Your Complete Name: ")
    yourAccountNumber = input("Select Your New Account Number: ")
    if yourAccountNumber in accountsDictionary:
        print("\n Uh-oh! The Provided Account Number Already Exists. Kindly Select a Unique One.")
        return
    password = input("Set a password for your account: ")
    accountsDictionary[yourAccountNumber] = {"completeName": completeName, "totalBalance": 0.0, "OverallTransactions": [], "password": password}
    print(f"\n🎉 The Account has now been created for {completeName} with Account # : {yourAccountNumber}.\n")

def depositMoneyInAcc():
    print("\n*** Deposit Money Into Your ACCOUNT ***")
    yourAccountNumber = input("Please Type Your Complete Account # Here: ")

    if yourAccountNumber not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found in Valid Accounts List! Please Try Again with the Correct one.\n")
        return
    try:
        totalAmount = float(input("Enter the amount you want to deposit in: $"))
        if totalAmount <= 0:
            print("\n❌ Invalid amount! Please enter a valid positive number.\n")
            return
        accountsDictionary[yourAccountNumber]["totalBalance"] += totalAmount
        accountsDictionary[yourAccountNumber]["OverallTransactions"].append(f"Deposited: +${totalAmount:.2f}")
        print(f"\n✅ The amount has been deposited successfully! ${totalAmount:.2f} has been added to your account.\n")
    except ValueError:
        print("\n⚠️ Oops! That wasn't a valid amount. Please enter a number.\n")

def withdraw_money():
    print("\n*** Withdraw Money ***")
    yourAccountNumber = input("Please Type Your Complete Account # Here: ")
    if yourAccountNumber not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found in Valid Accounts List! Please Try Again with the Correct one.\n")
        return
    try:
        totalAmount = float(input("Enter the amount to withdraw: $"))
        if totalAmount <= 0:
            print("\n❌ Invalid amount! Please enter a valid positive number.\n")
            return
        if accountsDictionary[yourAccountNumber]["totalBalance"] < totalAmount:
            print(f"\n⚠️ Insufficient funds! You have only ${accountsDictionary[yourAccountNumber]['totalBalance']:.2f} in your account.\n")
            return
        accountsDictionary[yourAccountNumber]["totalBalance"] -= totalAmount
        accountsDictionary[yourAccountNumber]["OverallTransactions"].append(f"Withdrew: -${totalAmount:.2f}")
        print(f"\n✅ Withdrawal successful! ${totalAmount:.2f} withdrawn from your account.\n")
    except ValueError:
        print("\n⚠️ Oops! That wasn't a valid amount. Please enter a number.\n")

def check_balance():
    print("\n*** Check Your Balance Here ***")
    yourAccountNumber = input("Please Type Your Complete Account # Here: ")
    if yourAccountNumber not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found in Valid Accounts List! Please Try Again with the Correct one.\n")
        return
    balance = accountsDictionary[yourAccountNumber]["totalBalance"]
    print(f"\nYour current balance is: ${balance:,.2f}\n")

def view_transactions():
    print("\n*** View Your Transaction History ***")
    yourAccountNumber = input("Please Type Your Complete Account # Here: ")
    if yourAccountNumber not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found! Please Try Again with the Correct one.\n")
        return
    transactions = accountsDictionary[yourAccountNumber]["OverallTransactions"]
    if not transactions:
        print("\nNo transactions found for the provided account.\n")
    else:
        print("\nTransaction history:")
        for transaction in transactions:
            print(f"- {transaction}")
        print()

def change_password():
    print("\n*** Change Your Account Password ***")
    yourAccountNumber = input("Enter Your Account Number: ")
    if yourAccountNumber not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found! Please Try Again with the Correct one.\n")
        return
    oldPassword = input("Enter your old password: ")
    if accountsDictionary[yourAccountNumber]["password"] != oldPassword:
        print("\n❌ Incorrect password. Please try again.\n")
        return
    newPassword = input("Enter your new password: ")
    accountsDictionary[yourAccountNumber]["password"] = newPassword
    print("\n✅ Your password has been successfully changed.\n")

def transfer_funds():
    print("\n*** Transfer Funds ***")
    fromAccount = input("Enter the Account Number you want to transfer money from: ")
    if fromAccount not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found! Please Try Again with the Correct one.\n")
        return
    toAccount = input("Enter the Account Number you want to transfer money to: ")
    if toAccount not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found! Please Try Again with the Correct one.\n")
        return
    try:
        amount = float(input("Enter the amount you want to transfer: $"))
        if amount <= 0:
            print("\n❌ Invalid amount! Please enter a positive number.\n")
            return
        if accountsDictionary[fromAccount]["totalBalance"] < amount:
            print(f"\n⚠️ Insufficient funds! You have only ${accountsDictionary[fromAccount]['totalBalance']:.2f} in your account.\n")
            return
        accountsDictionary[fromAccount]["totalBalance"] -= amount
        accountsDictionary[toAccount]["totalBalance"] += amount
        accountsDictionary[fromAccount]["OverallTransactions"].append(f"Transferred: -${amount:.2f} to Account {toAccount}")
        accountsDictionary[toAccount]["OverallTransactions"].append(f"Received: +${amount:.2f} from Account {fromAccount}")
        print(f"\n✅ Transfer successful! ${amount:.2f} has been transferred from Account #{fromAccount} to Account #{toAccount}.\n")
    except ValueError:
        print("\n⚠️ Oops! That wasn't a valid amount. Please enter a number.\n")

def login():
    print("\n*** Login to Your Account ***")
    yourAccountNumber = input("Enter Your Account Number: ")
    if yourAccountNumber not in accountsDictionary:
        print("\n🚫 The Provided Account NOT Found! Please Try Again with the Correct one.\n")
        return None
    password = input("Enter your password: ")
    if accountsDictionary[yourAccountNumber]["password"] != password:
        print("\n❌ Incorrect password. Please try again.\n")
        return None
    print(f"\n✅ Login successful! Welcome back, {accountsDictionary[yourAccountNumber]['completeName']}!\n")
    return yourAccountNumber

def simpleBankingSystem():
    print("\n" + "="*50)
    print("   🌟 Welcome to The Simple Banking System! 🌟")
    print("  💰 Your Personal Money Management System 💰")
    print("="*50 + "\n")
    while True:
        print("       🔑 Your Banking Portal ")
        print("\nHere are the Services We Offer to Our Valued Customers:\n")
        print("  1. 💼 Create Account (C) ")
        print("  2. 🚪 Login (L) ")
        print("  3. 💵 Deposit Money (D) ")
        print("  4. 💳 Withdraw Money (W) ")
        print("  5. 💰 Check Balance (B) ")
        print("  6. 🔍 View Transactions (T) ")
        print("  7. 🔑 Change Password (P) ")
        print("  8. 💸 Transfer Funds (F) ")
        print("  9. 🚪 Exit (E)\n")
        print("="*50 + "\n")
        userSelection = input("Type the service option you want to avail among the (C, L, D, W, B, T, P, F, E) letters : \n C for Create | L for Login | D for Deposit | W for Withdraw | B for Balance Check | T for Transaction History | P for Change Password | F for Fund Transfer | E for Exit.\n\n ")
        if userSelection == 'C':
            createNewAccount()
        elif userSelection == 'L':
            login()
        elif userSelection == 'D':
            depositMoneyInAcc()
        elif userSelection == 'W':
            withdraw_money()
        elif userSelection == 'B':
            check_balance()
        elif userSelection == 'T':
            view_transactions()
        elif userSelection == 'P':
            change_password()
        elif userSelection == 'F':
            transfer_funds()
        elif userSelection == 'E':
            print("🚪 Thank you for using our Banking System! Have a GOOD DAY!\n")
            break
        else:
            print("⚠️ Oops! The chosen option does not exist. Please try among the (C, D, W, B, T, P, F, E) letters.")
if __name__ == "__main__":
    simpleBankingSystem()
```
